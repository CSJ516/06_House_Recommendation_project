## 필터링

```python
def full(input_rent,input_deposit): 
    cur.execute(f"SELECT address, rent, deposit::integer, year::integer FROM officetels \
                WHERE rent='{input_rent}'\
                and deposit<='{input_deposit}'") 
    row = cur.fetchall() 
    
    return row
```

```python
def month(input_rent,input_deposit,input_pay): 
    cur.execute(f"SELECT address, rent, deposit::integer, pay::integer, year::integer FROM youn \
                WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                and pay <= '{input_pay}'") 
    row = cur.fetchall() 
    
    return row
```

```python
def condition(input_rent):
    if input_rent =='전세':
        input_deposit = input("deposit: ")
        return full(input_rent,input_deposit)
    elif input_rent == '월세':
        input_deposit = input("deposit: ")
        input_pay = input("pay: ")
        return month(input_rent,input_deposit,input_pay)
```

```python
#좌표
def month(input_rent,input_deposit,input_pay): 
    cur.execute(f"SELECT * \
                FROM \
                    (select address, rent, deposit::integer, pay::integer, latitude::float, longitude::float, recent from youn \
                    WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                    and pay <= '{input_pay}' and recent = 1 order by pay desc) as r \
                LIMIT 20") 
    row = cur.fetchall()
    row = pd.DataFrame(row)

    maps = folium.Map(location=[37.5642135, 127.0016985], zoom_start=12)
    for i in range(len(row)): 
        if (__name__ == "__main__") :
            origin          = '"'+ f"{row.iloc[i][4]},{row.iloc[i][5]}" + '"'
            destination     = "37.4862842,127.0204476"
            mode            = "transit"
            departure_time  = "now"
            key             = gmaps_key

            url = "https://maps.googleapis.com/maps/api/directions/json?origin="+ origin \
                    + "&destination=" + destination \
                    + "&mode=" + mode \
                    + "&departure_time=" + departure_time\
                    + "&language=ko" \
                    + "&key=" + key

            request         = urllib.request.Request(url)
            context         = ssl._create_unverified_context()
            response        = urllib.request.urlopen(request, context=context)
            responseText    = response.read().decode('utf-8')
            responseJson    = json.loads(responseText)

            with open("./Agent_Transit_Directions.json","w") as rltStream :
                json.dump(responseJson,rltStream)


        if ( __name__ == "__main__" ) :

            wholeDict = None
            with open("./Agent_Transit_Directions.json","r") as transitJson :
                wholeDict = dict(json.load(transitJson))

            path            = wholeDict["routes"][0]["legs"][0]
            duration_sec    = path["duration"]["value"]
            start_geo       = path["start_location"]
            end_geo         = path["end_location"]
            
            minute = str(round(int(duration_sec)/60,1)) + '분'
            print(minute) 
            print(end_geo)
#             out = f'{row.iloc[i][0]}  {\n}  {minute}'
            out = "Address : {} {} time : {}".format(row.iloc[i][0], "\n", minute, )
            popup = folium.Popup(out, max_width=2650)
            folium.Marker([row.iloc[i][4], row.iloc[i][5]], popup=popup,icon=folium.Icon(icon='cloud')).add_to(maps)

    return maps
```
